```python
import pandas as pd
pre_dict=pd.read_excel("./실험4_용어사전.xlsx")
```

```python
health_dict=pre_dict[pre_dict['SNT_STD_LCLS_NM(대)']=='정보/통신']['CNFRMTN_KOR_KYWD']
```

```python
health_key=list(health_dict)
```

```python
len(health_key)
```

```python
# 새로운 리스트
import re
list_par = []

# 텍스트를 가지고 있는 리스트

for i in health_key:
    text = re.sub('[a-zA-Z0-9]',' ',i).strip()
    # 빈 리스트는 제거.
    if(text != ''):
        list_par.append(text)
```

```python
len(list_par)
```

```python
tokens=sorted(list_par, key= len, reverse=False)
```

```python
import torch
from torch import nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np

import tokenization_kobert

model_version = 'monologg/kobert'
tokenizer = tokenization_kobert.KoBertTokenizer.from_pretrained(model_version)
```

```python
for i in range(len(tokens)):
    print(tokens[i])
    tokenizer.add_tokens(tokens[i])
    print(tokenizer.convert_tokens_to_ids(tokens[i]))
```
